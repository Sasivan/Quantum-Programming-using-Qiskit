```python
# Verify Qiskit 2.x and AerSimulator
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

print("Qiskit AerSimulator test (Qiskit 2.x)")

# simple circuit: create superposition and measure
qc = QuantumCircuit(1, 1)
qc.h(0)
qc.measure_all()

print("Circuit:")
print(qc.draw())

# run on AerSimulator
sim = AerSimulator()
# compile/transpile for the simulator
tqc = transpile(qc, sim)
job = sim.run(tqc, shots=1024)
result = job.result()
counts = result.get_counts()
print("Counts:", counts)
```

```python
# Task 1: Change number of shots
# Run with shots=10, 1000, and 10000. Compare results.
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

for shots in [10, 1000, 10000]:
    qc = QuantumCircuit(1, 1)
    qc.h(0)
    qc.measure(0, 0)
    sim = AerSimulator()
    tqc = transpile(qc, sim)
    job = sim.run(tqc, shots=shots)
    result = job.result()
    counts = result.get_counts()
    print(f"Shots: {shots}, Counts: {counts}")
```

```python
# Task 2: Two-qubit superposition
# Apply H to both qubits. Expect outcomes: 00, 01, 10, 11 (~25% each)

qc = QuantumCircuit(2, 2)
qc.h(0)
qc.h(1)
qc.measure([0, 1], [0, 1])
sim = AerSimulator()
tqc = transpile(qc, sim)
job = sim.run(tqc, shots=1024)
result = job.result()
counts = result.get_counts()
print("Counts:", counts)
```

```python
# Task 3: Entangled state (Bell state)
# Apply H to qubit 0, then CX from 0 to 1, measure all

qc = QuantumCircuit(2, 2)
qc.h(0)
qc.cx(0, 1)
qc.measure([0, 1], [0, 1])
sim = AerSimulator()
tqc = transpile(qc, sim)
job = sim.run(tqc, shots=1024)
result = job.result()
counts = result.get_counts()
print("Counts:", counts)
```

```python
# Task 4: Try an X gate
# Replace H with X. Expected: always 1

qc = QuantumCircuit(1, 1)
qc.x(0)
qc.measure(0, 0)
sim = AerSimulator()
tqc = transpile(qc, sim)
job = sim.run(tqc, shots=1024)
result = job.result()
counts = result.get_counts()
print("Counts:", counts)
```

```python
# Task 5: Three-qubit superposition (challenge)
# Apply H to all 3 qubits. Expect 8 outcomes (~12.5% each)

qc = QuantumCircuit(3, 3)
qc.h([0, 1, 2])
qc.measure([0, 1, 2], [0, 1, 2])
sim = AerSimulator()
tqc = transpile(qc, sim)
job = sim.run(tqc, shots=1024)
result = job.result()
counts = result.get_counts()
print("Counts:", counts)
```
